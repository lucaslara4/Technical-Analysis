### Moving Average Crossover Trading Strategy

#### How It Works

The strategy utilizes two moving averages:

- **Short-term Moving Average ($MA_{\text{short}}$)**: calculated over a smaller window $n_s$ (e.g., 50 days).  
- **Long-term Moving Average ($MA_{\text{long}}$)**: calculated over a larger window $n_l > n_s$ (e.g., 200 days).

Formally:

$$
MA_{\text{short}}(t) = \frac{1}{n_s} \sum_{k=0}^{n_s-1} P_{t-k}
$$

$$
MA_{\text{long}}(t) = \frac{1}{n_l} \sum_{k=0}^{n_l-1} P_{t-k}
$$

Where $P_t$ is the closing price at time $t$.

#### Trading Signals

- **Buy Signal ("Golden Cross")**: occurs when $MA_{\text{short}}(t)$ crosses above $MA_{\text{long}}(t)$.  
  This signal suggests growing bullish momentum and a likely shift from a downtrend to an uptrend. It indicates a potential entry point for buying.

- **Sell Signal ("Death Cross")**: occurs when $MA_{\text{short}}(t)$ crosses below $MA_{\text{long}}(t)$.  
  This signal suggests increasing bearish momentum and a shift from an uptrend to a downtrend. It is typically a cue to exit or sell positions.

##### Signal Logic in Math

$$
\begin{aligned}
\text{Buy Signal:} \quad & MA_{\text{short}}(t-1) \leq MA_{\text{long}}(t-1) \quad \text{and} \quad MA_{\text{short}}(t) > MA_{\text{long}}(t) \\
\text{Sell Signal:} \quad & MA_{\text{short}}(t-1) \geq MA_{\text{long}}(t-1) \quad \text{and} \quad MA_{\text{short}}(t) < MA_{\text{long}}(t)
\end{aligned}
$$

#### Exponential Moving Average (EMA) Enhancement

Instead of using simple moving averages, this strategy uses exponential moving averages (EMA) defined as:

$$
EMA_t = \alpha \cdot P_t + (1 - \alpha) \cdot EMA_{t-1}
$$

where smoothing factor $ \alpha = \frac{2}{n+1} $ and $n$ is the window size.

The EMA weights recent prices more heavily, allowing the indicator to respond faster to price changes than the SMA.

#### Why EMA is Preferred

- **Faster response** to recent price changes.  
- **Reduced lag**, enabling earlier detection of trend changes.  
- **Maintains smoothness** while being more sensitive, ideal for trading signals.

Using EMA improves the timeliness and reliability of buy and sell signals in this crossover strategy.

#### Example Log Decision

Every time a crossover event is detected, the strategy prints the date, closing price, EMAs, signal type (Golden Cross or Death Cross), and action (Buy or Sell).

# Code For a single security

```python
%matplotlib inline

import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt
import datetime

# Parameters
ticker = "SOXL"
short_window = 50
long_window = 200

# Download historical data (last year)
data = yf.download(ticker, period="1y", interval="1d", auto_adjust=True, progress=False)

# Diagnostic: print DataFrame columns to check for duplicates or multi-index
print("Data columns:", data.columns)

if data.empty:
    print(f"No data found for {ticker}. Please check the ticker or data source.")
else:
    print(f"Data for {ticker} downloaded with {len(data)} rows.")

    # Calculate Exponential Moving Averages (EMA)
    data['Short_EMA'] = data['Close'].ewm(span=short_window, adjust=False).mean()
    data['Long_EMA'] = data['Close'].ewm(span=long_window, adjust=False).mean()

    in_position = False
    trades = []

    print("Date       | Close   | Short_EMA | Long_EMA | Signal       | Action")
    print("---------------------------------------------------------------")

    for i in range(1, len(data)):
        row = data.iloc[i]
        prev = data.iloc[i - 1]
        date = row.name.date()

        # Safely get single scalar values if they are Series
        def safe_scalar(val):
            if isinstance(val, pd.Series):
                return val.iloc[0]
            return val

        short_ema = safe_scalar(row['Short_EMA'])
        long_ema = safe_scalar(row['Long_EMA'])
        prev_short_ema = safe_scalar(prev['Short_EMA'])
        prev_long_ema = safe_scalar(prev['Long_EMA'])
        close_price = safe_scalar(row['Close'])

        action = ""

        if pd.notna(short_ema) and pd.notna(long_ema):
            if float(short_ema) > float(long_ema) and float(prev_short_ema) <= float(prev_long_ema):
                action = "BUY"
                in_position = True
                print(f"{date} | {float(close_price):.2f} | {float(short_ema):.2f} | {float(long_ema):.2f} | Golden Cross | {action}")
                trades.append({'date': row.name, 'type': 'BUY', 'price': float(close_price)})

            elif float(short_ema) < float(long_ema) and float(prev_short_ema) >= float(prev_long_ema):
                if in_position:
                    action = "SELL"
                    in_position = False
                    print(f"{date} | {float(close_price):.2f} | {float(short_ema):.2f} | {float(long_ema):.2f} | Death Cross  | {action}")
                    trades.append({'date': row.name, 'type': 'SELL', 'price': float(close_price)})

    print("\nTrade Log:")
    for t in trades:
        print(f"{t['date'].date()} - {t['type']} at {t['price']:.2f}")

    # Plot the price and EMAs with updated legend labels
    plt.figure(figsize=(14, 7))
    plt.plot(data['Close'], label='Close Price')
    plt.plot(data['Short_EMA'], label='50-day short EMA')
    plt.plot(data['Long_EMA'], label='200-day long EMA')

    buy_dates = [t['date'] for t in trades if t['type'] == 'BUY']
    buy_prices = [t['price'] for t in trades if t['type'] == 'BUY']
    sell_dates = [t['date'] for t in trades if t['type'] == 'SELL']
    sell_prices = [t['price'] for t in trades if t['type'] == 'SELL']

    plt.scatter(buy_dates, buy_prices, marker='^', color='green', s=100, label='Buy Signal')
    plt.scatter(sell_dates, sell_prices, marker='v', color='red', s=100, label='Sell Signal')

    plt.title(f"{ticker} Price and Exponential Moving Averages")
    plt.legend()
    plt.grid(True)
    plt.show()

# Main analysis and concise recommendation (summary bullet points)
today = datetime.datetime.today().date()
if trades:
    last_trade = trades[-1]
    if last_trade['type'] == 'BUY':
        recommendation = (
            "- Last trade was a Buy due to a Golden Cross (50-day EMA crossed above 200-day EMA).\n"
            "- This is a bullish indicator suggesting an upward price trend.\n"
            "- Maintaining or increasing position is recommended while monitoring market conditions.\n"
            "- Watch for any reversal signals or Death Cross for exit."
        )
    else:
        recommendation = (
            "- Last trade was a Sell triggered by a Death Cross (50-day EMA crossed below 200-day EMA).\n"
            "- This is a bearish indicator signaling potential downward momentum.\n"
            "- Staying out of the position is prudent until a new buy signal occurs.\n"
            "- Remain attentive to future Golden Cross for potential re-entry."
        )
else:
    recommendation = (
        "- No trades executed in the period analyzed.\n"
        "- Monitor for EMA cross signals: Golden Cross to enter, Death Cross to exit.\n"
        "- Use this strategy to guide timing decisions based on trend changes."
    )

print(f"\nMain Analysis as of {today}:\n{recommendation}")

```
